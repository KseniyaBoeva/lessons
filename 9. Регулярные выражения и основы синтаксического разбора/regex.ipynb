## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'А222ВС96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re

def check_car_id(car_id):
    number = re.findall(r'^[АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2}', car_id)
    region = re.findall(r'\d{2,3}$', car_id) 
    rez = number + region
    return rez
    
car_id       = input('Введите номер: ')
car_id_parts = check_car_id(car_id)
if len(car_id_parts) <= 1:
    print('Номер не валиден')
else:
    print(f'Номер {car_id_parts[0]} валиден. Регион: {car_id_parts[1]}')
````

## Задание 2

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат: 
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
import re
def delete_repeats(some_string):
    result = re.split(r'\W{1,}', some_string)
    new_string = ''
    old_word   = ''
    for word in result:
        if old_word == word:
            continue
            
        new_string = new_string + " " + word
        old_word   = word
        
    return new_string

some_string   = input('Введите строку для форматирования: ')
format_string = delete_repeats(some_string)
print(f'Результат: {format_string}')
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
import re
def get_acronim(some_words):    
    result = re.findall(r'\b\w', some_words.title() )
    result_2 = ''.join(result)
    return result_2

some_words = input('Введите слово: ')
acronim = get_acronim(some_words)
print(f'Результат: {acronim}')
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
import re
def categorize_emails(emails):
    domens = re.findall(r'(?<=@)\w{1,}\.\w{2,}', emails)
    distribute     = {}
    for domen in domens:
        if domen in distribute:
            distribute[domen] = distribute[domen] + 1
        else:
            distribute[domen] = 1
            
    return distribute

emails = input('Введите email адреса: ')
domen_list = categorize_emails(emails)
print(f'Результат: {domen_list}')
````
